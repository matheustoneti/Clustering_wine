```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import cluster
```

```python
dados = pd.read_csv('wine-clustering.csv')
dados.head ()
```

```python
dados_cluster = dados[['Total_Phenols', 'Color_Intensity','Hue']]
sns.pairplot(dados_cluster, hue='Hue')
```

```python
fig, axe = plt.subplots (figsize =(8,5))
axe = sns.scatterplot (data=dados_cluster, x='Total_Phenols', y= 'Color_Intensity')
```

```python
kmeans = cluster.KMeans(3)
clusters = kmeans.fit_predict(dados_cluster)
pd.value_counts(clusters)
```

```python
x_center, y_center = kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1]
fig, (axe1,axe2) = plt.subplots(1,2, figsize=(14,5))
sns.scatterplot (data = dados_cluster, x='Total_Phenols', y='Color_Intensity', hue= clusters, ax=axe1)
axe1.scatter(x=x_center, y=y_center, s=80, color='red')
sns.scatterplot( data=dados_cluster, x='Total_Phenols', y= 'Color_Intensity', hue='Hue', ax=axe2)
```
